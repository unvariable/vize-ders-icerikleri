```python
frame=pd.DataFrame(np.arange(12).reshape((4,3)),#dataframe açma
                  index=[['a','a','b','b'],[1,2,1,2]],
                   columns=[['Ankara','Ankara','İstanbul'],['Kırmızı','Mavi','Sarı']])
frame
```

```python
frame.index.names=['Anahtar1','Anahtar2'] #Anahtar isimlendirme
frame.columns.names=['Sehir','Renk']
frame
```

```python
frame['Ankara']
```

```python
#Sütunların yerini değiştirme
frame.swaplevel('Anahtar1','Anahtar2')
```

```python
frame
```

```python
frame.sort_index(level=1)#sıralama yapma
```

```python
frame.swaplevel(0,1).sort_index(level=1)
```

```python
frame.groupby(level='Anahtar2').sum()#Seviye bazlı toplama işlemi
#                                     Anahtar 2 değerlerini topluyor
#                                     Eski versiyonda frame.sum(level='Anahtar2') olarak kullanılıyor.
```
