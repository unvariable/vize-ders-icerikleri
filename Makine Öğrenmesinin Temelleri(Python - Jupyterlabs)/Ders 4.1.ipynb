```python
import pandas as pd
import numpy as np
```

```python
left1 = pd.DataFrame({'key':['a','b','a','a','b','c'],
                     'value':range(6)})
left1
```

```python
right1=pd.DataFrame({'group_val':[3.5,7]},index=['a','b'])
right1
```

```python
pd.merge(left1,right1,left_on='key',right_index=True)#iki dataframe'yi birleştirmek
```

```python
pd.merge(left1,right1,left_on='key',right_index=True,how='outer')#how='outer' anahtar sözcüğü kullanımı
```

```python
pd.merge(left1,right1,left_on='key',right_index=True,how='inner')#default olarak how='inner'
```

```python
lefth=pd.DataFrame({'key1':['İstanbul','İstanbul','İstanbul','Ankara','Ankara'],
                   'key2':[2000,2001,2002,2001,2002],
                   'data':np.arange(5.)})
lefth
```

```python
righth=pd.DataFrame(np.arange(12).reshape((6,2)),
                   index=[['Ankara','Ankara','İstanbul','İstanbul','İstanbul','İstanbul'],
                         [2001,2000,2000,2000,2001,2002]],
                   columns=['Olay1','Olay2'])
righth
```

```python
pd.merge(lefth,righth,left_on=['key1','key2'],right_index=True)
```
